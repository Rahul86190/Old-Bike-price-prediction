ML is a mathematical based algoritham / it understand only number 0-9 . not understang  alphabets anf alphanumerical values

inter encoding = convert into (1,2,3,4,5,6,7) 

one hot encoding = (0,1)

```python
df["owner"].value_counts()
```

owner name is alphabetical values . need to convert into numbers 

```python
dt={"First Owner":1,"Second Owner":2,"Third Owner":3,"Fourth Owner Or More":4}
```

```python
dt
```

to convert into numerical here we use map() function . this will change the value accordingly the dictionary we made and set values (1,2,3,4) to keys( First Owner, Second Owner ,Third Owner,Fourth Owner Or More)

```python
df["owner"].map(dt)
```

This is also known as ordinalencoding  = assign intergers in a meaningfull order.

But in label encoding , we can assign any integer no without any meaning

```python
df["owner"]=df["owner"].map(dt)
```

```python
df
```

```python
df["brand"].value_counts()
```

```python
brand_ls=list(df["brand"].value_counts().head(16).keys())
brand_ls
```

```python
df2=df[df["brand"].isin(brand_ls)]
df2
```

```python
l2=list[df2["brand"].value_counts().keys()]
l2
```

```python
# dictionary comprehension
brand_dict={brand:i for i ,brand in enumerate(list(df2["brand"].value_counts().keys()),start=1)}
print(brand_dict)
df2["brand"]=df2["brand"].map(brand_dict)
```

```python
df2
```

Feature selection method == deleting column

```python
df2.drop(["bike_name","city"],axis=1,inplace=True)
```

```python
df2.head(4)
```

x are features and y are target

```python
x=df2.drop("price",axis=1)
x
```

```python
y=df2[["price"]]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(x,y,test_size=0.2)   #  test size =20% so training size=80%
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)   #  test size =20% so training size=80%
```

```python
x_train
```

Linear regression model tranning

```python
from sklearn.linear_model import LinearRegression   # algorithm import
```

LinearRgression in a class

```python
lnr=LinearRegression()
```

```python
lnr.fit(x_train,y_train)   # use to train
```

```python
lnr.score(x_train,y_train)*100   # 75% accuracy  . it is  score of seen data ( train data)
```

```python
# unseen data score (test data)
lnr.score(x_test,y_test)*100
```

prediction

```python
y_test
```

```python
prediction=lnr.predict(x_test)
```

```python
y_test["Prediction"]=prediction
```
